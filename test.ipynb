```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('test_file.csv')

# Ensure the 'Date' column is the datetime type
df['Date'] = pd.to_datetime(df['Date'])

# Set the 'Date' column as the DataFrame's index
df.set_index('Date', inplace=True)

# Iterate over each column (stock) in the DataFrame
for column in df.columns:
    # Plot the stock's historical data
    plt.figure(figsize=(10, 6))
    df[column].plot(title=f"Time Series for {column} Stock")
    plt.xlabel('Date')
    plt.ylabel('Stock Price')
    plt.show()
```

```python


# Calculate the percent change and then the cumulative product to reflect overall growth
df_percent_change = df.pct_change().add(1).cumprod().sub(1)

# Plotting
plt.figure(figsize=(14, 7))  # Set the figure size for better readability
for column in df_percent_change.columns:
    plt.plot(df_percent_change.index, df_percent_change[column], label=column)

plt.title('Percent Change in Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Percent Change (%)')
plt.legend(title='Stock Ticker', bbox_to_anchor=(1.05, 1), loc='upper left')  # Place legend outside the plot
plt.tight_layout()
plt.grid(True)  # Add a grid for better readability of percent changes
plt.show()
```

```python


# Input start and end date
start_date = input("Enter the start date (YYYY-MM-DD): ")
end_date = input("Enter the end date (YYYY-MM-DD): ")

# Convert user input to datetime
start_date = pd.to_datetime(start_date)
end_date = pd.to_datetime(end_date)

# Filter the DataFrame based on the input dates
filtered_df = df.loc[start_date:end_date]

# Calculate the percent change and then the cumulative product to reflect overall growth
df_percent_change = filtered_df.pct_change().add(1).cumprod().sub(1)

# Plotting
plt.figure(figsize=(14, 7))  # Set the figure size for better readability
for column in df_percent_change.columns:
    plt.plot(df_percent_change.index, df_percent_change[column], label=column)

plt.title('Percent Change in Stock Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Percent Change (%)')
plt.legend(title='Stock Ticker', bbox_to_anchor=(1.05, 1), loc='upper left')  # Place legend outside the plot
plt.tight_layout()
plt.grid(True)  # Add a grid for better readability of percent changes
plt.show()
```
